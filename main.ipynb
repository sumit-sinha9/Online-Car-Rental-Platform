```python
from car_rental import CarRental
from customer import Customer

def main():
    shop = CarRental(20)
    customer = Customer()

    while True:
        print("""
        ====== Car Rental Service =======
        1. Display available cars
        2. Rent a car
        3. Return a car
        4. Exit
        """)

        choice = input("Enter choice: ")

        try:
            choice = int(choice)
        except ValueError:
            print("That's not an int!")
            continue

        if choice == 1:
            shop.display_stock()

        elif choice == 2:
            customer.rentalTime = shop.rent_car(customer.request_car(), customer.carModel, customer.carCategory)
            customer.rentalBasis = customer.carCategory

        elif choice == 3:
            customer.bill = shop.return_car(customer.return_car())
            customer.rentalBasis, customer.rentalTime, customer.cars, customer.carModel, customer.carCategory = 0, 0, 0, 0, 0

        elif choice == 4:
            break
        else:
            print("Invalid input. Please enter a number between 1-4.")

    print("Thank you for using the car rental service!")

if __name__ == "__main__":
    main()
```
